```python
#import numpy
import numpy as np
```

```python
#masukan input
#input 0 = produksi banyak (PB)
#input 1 = produksi kecil (PK)

#kombinasi input = 0,1 output 1
#kombinasi input = 1,2 output 0
#kombinasi input = 0,2 output 1

#output 0 = produksi besar
#output 1 = produksi kecil

#inputkan value
inputFeatures = np.array([[0,0],[0,1],[1,0],[1,1]])
targetOutput = np.array([[0],[0.3],[0.4],[1]])
weights = np.array([[0.1],[0.2]])
```

```python
#set learning rate
lr = 2
```

```python
#fungsi simgmoid
def sigmoid(x):
  return 1/(1+np.exp(-x))
  
#derivasi
def sigmoid_der(x):
  return sigmoid(x)*(1-sigmoid(x))
```
